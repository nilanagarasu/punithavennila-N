```python
from pandas import DataFrame
data = {'M01AB':[0.00,0.00, np.nan, 0.95],
		'M01AE': [0.67,0.00,0.01, np.nan],
		'N02BA':[np.nan,1.0,0.0,0.0]
		}
df = DataFrame(data, columns = ['M01AB', 'M01AE', 'N02BA'])
print("Original DataFrame:\n", df)
print("\nDescriptive statistics of M01AE:\n")
stats = df['M01AE'].describe()
print(stats)
```

```python
from pandas import DataFrame
data = {'M01AB':[0.00,0.00, np.nan, 0.95],
		'M01AE': [0.67,0.00,0.01, np.nan],
		'N02BA':[np.nan,1.0,0.0,0.0]
		}
df = DataFrame(data, columns = ['M01AB', 'M01AE', 'N02BA'])
print("Original DataFrame:\n", df)
print("\nDescriptive statistics of whole dataframe:\n")
stats = df.describe(include = 'all')
print(stats)
```

Creating Visualizations for Data Exploration:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
n_rows = 1000
data = {
    'Product': np.random.choice(['M01AB','M01AE','N02BA','N02BE'], size=n_rows),
    'datum': np.random.choice(['1/2/2014 08:00','1/2/2014 09:00'], size=n_rows),
    'Year': np.random.randint(500, 1500, size=n_rows),
    'Hour': np.random.randint(20, 50, size=n_rows),
}

df = pd.DataFrame(data)
product_quantity = df.groupby('Product')['Hour'].sum()
product_quantity.plot(kind='bar', color='skyblue')
plt.title('Bar Plot: Total Hour by Product')
plt.xlabel('Product')
plt.ylabel('Total Hour')
plt.show()
df['Year'].plot(kind='hist', bins=20, edgecolor='black', color='lightgreen')
plt.title('Histogram: Distribution of Prices')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
df.plot(kind='scatter', x='Year', y='Hour', alpha=0.5, color='coral')
plt.title('Scatter Plot: Year vs. Hour')
plt.xlabel('Year')
plt.ylabel('Hour')
plt.show()

```

Correlation and Data Relationships:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(20)
n_rows = 1000
data = {
    'Product': np.random.choice(['M01AB','M01AE','N02BA','N02BE'], size=n_rows),
    'datum': np.random.choice(['1/2/2014 08:00','1/2/2014 09:00'], size=n_rows),
    'Year': np.random.randint(500, 1500, size=n_rows),
    'Hour': np.random.randint(20, 50, size=n_rows),
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
